```python
df[['First Name','Last Name']] =df['Name'].str.split(" ",n=1,expand = True) 
df
```

```python
df = df.drop( columns = 'Name')
df
```

```python
col = df.pop("First Name")
df.insert(3,"First Name",col)
df
```

```python
col = df.pop("Last Name")
df.insert(4,"Last Name",col)
df
```

```python
df['First Name'] = df['First Name'].str.rstrip(',')
df
```

```python
df.rename(columns ={"PassengerId" : "PassengerID"},inplace = True)
df
```

```python
df['Survived'] = df['Survived'].astype(int)
df
```

```python
df['Survived'] = df['Survived'].map({0:'No', 1:'Yes'})
df
```

```python
df.info()
```

```python
df = df.drop(columns = 'Cabin')
df
```

```python
df = df.replace('NaN','')
df

```

```python
df['Embarked'] = df['Embarked'].fillna('')
df

```

```python
for x in df.index:
    if df.loc[x,'Embarked'] == '':
        df.drop(x,inplace=True)
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median(),inplace = True)
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
l_percentile = df['Fare'].quantile(0.01)
u_percentile = df['Fare'].quantile(0.99)
df['Fare'] = df['Fare'].clip(l_percentile,u_percentile )
df
```
